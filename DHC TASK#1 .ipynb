```python
# Import Libraries

import seaborn as sns 
import matplotlib.pyplot as plt
```

#  Step 1: Load 

```python
# Step 1: Load the dataset

iris = sns.load_dataset('iris')
```

# Step 2: Explore

```python
# Step 2: Explore the dataset

print("Dataset Shape:", iris.shape) 
print("Column Names:", iris.columns) 
print("\nFirst 5 Rows:\n", iris.head())
```

# Step 3: Visualizations

```python
# Scatter Plot - Sepal Length vs Sepal Width 

plt.figure(figsize=(6,4)) 
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris) 
plt.title("Sepal Length vs Sepal Width") 
plt.show()
```

```python
# Histogram - Petal Length Distribution 

plt.figure(figsize=(6,4)) 
sns.histplot(iris['petal_length'], kde=True, bins=20) 
plt.title("Distribution of Petal Length") 
plt.show()
```

```python
# Box Plot - Sepal Length by Species

plt.figure(figsize=(6,4)) 
sns.boxplot(x='species', y='sepal_length', data=iris) 
plt.title("Sepal Length Spread by Species") 
plt.show()
```
